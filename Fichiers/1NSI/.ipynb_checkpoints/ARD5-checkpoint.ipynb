# ARD5 - Modèle de Von Neumann

Le modèle de Von Neumann ou modèle SISD (*single instruction single data*) décrit l'organisation générale d'un ordinateur. 

# 1 - Composants

## 1.1 L'unité arithmétique et logique ALU (*arithmetic and logic unit*)

Aussi appelée unitée de traitement (*process unit*), c'est le composant permettant d'effectuer les opérations arithmétiques et booléennes fondamentales (+, -, *, /, AND, OR, NOT, XOR, etc.).

Il contient un certain nombre de registres (circuits mémoire) :
- des registres de données ;
- un registre spécial servant d'entrée et de sortie appelé **accumulateur** (sortie). 


## 1.2 L'unité de contrôle CU (*control unit*)

C'est le composant qui organise la communication entre les autres composants 
et en particulier qui met en oeuvre le cycle d'exécution des instructions.

Il contient deux registres : 
- le **registre instruction** ou **IR** (*instruction register*) stockant d'adresse mémoire de l'instruction en cours ;
- et le **pointeur d'instruction** ou **IP** (*instruction pointer*) stockant l'adresse mémire de la prochaine instruction.


Pris ensembles, ces deux composants forment le **processeur** ou **CPU** (*control and process unit*). Si de plus, ils sont regroupés sur une seule et même puce, on parle de **microprocesseur**.

## 1.3 La mémoire 

Attention : on ne parle pas ici du disque dur qui est considéré comme un périphérique. 

Dans la mémoire on stocke à la fois les données (variables) et les programmes (suites d'instructions que le CPU doit exécuter).

On distingue deux type de mémoires :

- la **mémoire volatile ou mémoire vive** : perd son contenu si on coupe l'alimentation, on l'utilise pour sa rapidité d'accès, en lecture comme en écriture. On parle aussi de **RAM** (*random access memory*)
- la mémoire non-volatile qui conserve son contenu même si on coupe l'alimentation. On utilise par exemple la **ROM** (*read-only memory*) pour stocker les instruction du programme permettant de démarrer la machine et toutes autres données statiques lui permettant de fonctionner.



## 1.4 périphériques d'entrée-sortie

Tous les autres composants sont considérés comme des ajouts à la machine. Ils sont connectés à celle-ci grâce à des **ports d'entrée-sortie (port IO)**. Voici une liste d'exemples : 

Périphériques d'entrée :
- clavier, souris ;
- manettes, lecteur de code barre ;
- microphone, scanner, appareil photo, webcam.

Périphériques de sortie :
- écran, vidéo-projecteur ;
- imprimante ;
- hauts parleurs, casque audio.

Les deux :
- lecteur graveur de CD / DVD / Blueray ;
- disques durs, clés USB, cartes SD ;
- cartes réseaux.

# 2 - Communications : les bus de données

On appelle **bus de données** un ensemble de $n$ fils (au sens des circuits) généralement $n=32 ou 64$ permettant de connecter entre-eux les différents composant de l'ordinateur.

- L'unité de contrôle est reliée à l'ALU, à la mémoire et aux ports IO.
- L'ALU est reliée à l'ALU et à la mémoire.
- Les ports IO peuvent avoir accès à la mémoire via un composant spcialisé : le **DMA** (*direct memory acces*)
